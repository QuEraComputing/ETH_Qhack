```python
# Using Squin
from typing import Any

from bloqade import squin, tsim
from bloqade.pyqrack import StackMemorySimulator
from bloqade.types import MeasurementResult, Qubit
from kirin.dialects.ilist import IList
# import tsim

# this will help us have return types for our methods that have more intuitive names
Register = IList[Qubit, Any]
Measurement = IList[MeasurementResult, Any]

# this function will help us visualize some circuits
def show_circuit(squin_kernel, h=400):
    @squin.kernel
    def _to_visualize():
        _ = squin_kernel()

    return tsim.Circuit(_to_visualize).diagram(height=h)


@squin.kernel
def bell_state() -> Measurement:
    qubits = squin.qalloc(2)
    squin.h(qubits[0])
    squin.cx(qubits[0], qubits[1])
    bits = squin.broadcast.measure(qubits)
    return bits

show_circuit(bell_state)

def build_stabilizer_supports(n):
    """
    Builds the stabilizer supports for a rotated n x n surface code.

    Returns:
        z_supports: tuple of tuples
        x_supports: tuple of tuples

    Data qubit convention:
        0   1   2
        3   4   5
        6   7   8
    """

    if n < 3:
        raise ValueError("n must be at least 3")

    if n % 2 == 0:
        raise ValueError("n must be odd: 3, 5, 7, ...")

    expected = (n**2 - 1) // 2

    def q(row, col):
        return row * n + col

    z_supports = []
    x_supports = []

    # =========================
    # Z STABILIZERS
    # =========================

    # Top border: horizontal pairs
    for col in range(0, n - 1, 2):
        z_supports.append((
            q(0, col),
            q(0, col + 1),
        ))

    # Z interior: 2x2 plaquettes with odd parity
    for row in range(n - 1):
        for col in range(n - 1):
            if (row + col) % 2 == 1:
                z_supports.append((
                    q(row, col),
                    q(row, col + 1),
                    q(row + 1, col),
                    q(row + 1, col + 1),
                ))

    # Bottom border: staggered horizontal pairs
    for col in range(1, n - 1, 2):
        z_supports.append((
            q(n - 1, col),
            q(n - 1, col + 1),
        ))

    # =========================
    # X STABILIZERS
    # =========================

    for row in range(n - 1):

        # Left border: staggered vertical pairs
        if row % 2 == 1:
            x_supports.append((
                q(row, 0),
                q(row + 1, 0),
            ))

        # X interior: 2x2 plaquettes with even parity
        for col in range(n - 1):
            if (row + col) % 2 == 0:
                x_supports.append((
                    q(row, col),
                    q(row, col + 1),
                    q(row + 1, col),
                    q(row + 1, col + 1),
                ))

        # Right border: vertical pairs
        if row % 2 == 0:
            x_supports.append((
                q(row, n - 1),
                q(row + 1, n - 1),
            ))

    if len(z_supports) != expected:
        raise RuntimeError(
            f"Z stabilizers generated: {len(z_supports)}, expected: {expected}"
        )

    if len(x_supports) != expected:
        raise RuntimeError(
            f"X stabilizers generated: {len(x_supports)}, expected: {expected}"
        )

    return tuple(z_supports), tuple(x_supports)

def surface_debug_text(n, z_supports, x_supports):
    """
    Creates a debug string instead of printing directly.
    More reliable in Jupyter: you can do print(debug_text).
    """

    lines = []

    lines.append("\nDATA QUBITS")
    lines.append("=" * 70)

    for row in range(n):
        line = "   ".join(f"{row * n + col:02d}" for col in range(n))
        lines.append(line)

    lines.append("\nZ STABILIZERS")
    lines.append("=" * 70)

    for i, support in enumerate(z_supports):
        lines.append(f"z_ancillas[{i:02d}] -> {list(support)}")

    lines.append("\nX STABILIZERS")
    lines.append("=" * 70)

    for i, support in enumerate(x_supports):
        lines.append(f"x_ancillas[{i:02d}] -> {list(support)}")

    lines.append("\nVISUAL MAP")
    lines.append("=" * 70)
    lines.append("D = data qubit, Z = Z stabilizer, X = X stabilizer\n")

    size = 2 * n + 1
    grid = [["." for _ in range(size)] for _ in range(size)]

    # Data qubits
    for row in range(n):
        for col in range(n):
            data_idx = row * n + col
            grid[2 * row + 1][2 * col + 1] = f"D{data_idx:02d}"

    # Z stabilizers
    z_counter = 0

    for col in range(0, n - 1, 2):
        grid[0][2 * col + 2] = f"Z{z_counter:02d}"
        z_counter += 1

    for row in range(n - 1):
        for col in range(n - 1):
            if (row + col) % 2 == 1:
                grid[2 * row + 2][2 * col + 2] = f"Z{z_counter:02d}"
                z_counter += 1

    for col in range(1, n - 1, 2):
        grid[2 * n][2 * col + 2] = f"Z{z_counter:02d}"
        z_counter += 1

    # X stabilizers
    x_counter = 0

    for row in range(n - 1):

        if row % 2 == 1:
            grid[2 * row + 2][0] = f"X{x_counter:02d}"
            x_counter += 1

        for col in range(n - 1):
            if (row + col) % 2 == 0:
                grid[2 * row + 2][2 * col + 2] = f"X{x_counter:02d}"
                x_counter += 1

        if row % 2 == 0:
            grid[2 * row + 2][2 * n] = f"X{x_counter:02d}"
            x_counter += 1

    for row in grid:
        lines.append("".join(f"{cell:>6}" for cell in row))

    return "\n".join(lines)
```

```python

def make_helpers(d, angle):
    z_supports, x_supports = build_stabilizer_supports(d)
    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    @squin.kernel
    def inizialize_qubits(data: Register, ancillas: Register) -> Measurement:
        for i in range(num_data):
            squin.reset(data[i])

        for i in range(num_ancillas):
            squin.reset(ancillas[i])

        for stab_idx in range(len(z_supports)):
            for data_idx in z_supports[stab_idx]:
                squin.cx(data[data_idx], ancillas[stab_idx])

        return squin.broadcast.measure(ancillas[:num_z])

    @squin.kernel  # must be a kernel, not plain fn
    def syndrome_round(data: Register, ancillas: Register) -> Measurement:
        for i in range(num_ancillas):
            squin.reset(ancillas[i])

        for i in range(num_x):
            squin.h(ancillas[num_z + i])

        for stab_idx in range(num_z):
            for data_idx in z_supports[stab_idx]:
                squin.cx(data[data_idx], ancillas[stab_idx])

        for stab_idx in range(num_x):
            for data_idx in x_supports[stab_idx]:
                squin.cx(ancillas[num_z + stab_idx], data[data_idx])

        for i in range(num_x):
            squin.h(ancillas[num_z + i])

        return squin.broadcast.measure(ancillas)

    @squin.kernel
    def physical_rotation(data:Register):
        for idx in range(0, num_data, d):
            squin.rz(angle, data[idx])

    return inizialize_qubits, syndrome_round, physical_rotation


```

```python
def make_helpers_noisy(d, angle_rad, p1=1e-3, p2=1e-3):
    """
    Noisy version of make_helpers.

    angle_rad:
        Physical angle in radians, because squin.rz(...) expects radians.

    p1:
        Depolarizing noise after 1-qubit gates: H, RZ.

    p2:
        Depolarizing noise after 2-qubit gates: CX.
    """

    z_supports, x_supports = build_stabilizer_supports(d)

    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    @squin.kernel
    def inizialize_qubits_noisy(data: Register, ancillas: Register) -> Measurement:
        # Prepare data qubits in |+>, not |0>
        for i in range(num_data):
            squin.reset(data[i])
            squin.h(data[i])
            if p1 > 0.0:
                squin.depolarize(p=p1, qubit=data[i])

        for i in range(num_ancillas):
            squin.reset(ancillas[i])

        for stab_idx in range(len(z_supports)):
            for data_idx in z_supports[stab_idx]:
                squin.cx(data[data_idx], ancillas[stab_idx])
                if p2 > 0.0:
                    squin.depolarize2(p2, data[data_idx], ancillas[stab_idx])

        return squin.broadcast.measure(ancillas[:num_z])

    @squin.kernel
    def syndrome_round_noisy(data: Register, ancillas: Register) -> Measurement:
        for i in range(num_ancillas):
            squin.reset(ancillas[i])

        for i in range(num_x):
            squin.h(ancillas[num_z + i])
            if p1 > 0.0:
                squin.depolarize(p=p1, qubit=ancillas[num_z + i])

        for stab_idx in range(num_z):
            for data_idx in z_supports[stab_idx]:
                squin.cx(data[data_idx], ancillas[stab_idx])
                if p2 > 0.0:
                    squin.depolarize2(p2, data[data_idx], ancillas[stab_idx])

        for stab_idx in range(num_x):
            for data_idx in x_supports[stab_idx]:
                squin.cx(ancillas[num_z + stab_idx], data[data_idx])
                if p2 > 0.0:
                    squin.depolarize2(p2, ancillas[num_z + stab_idx], data[data_idx])

        for i in range(num_x):
            squin.h(ancillas[num_z + i])
            if p1 > 0.0:
                squin.depolarize(p=p1, qubit=ancillas[num_z + i])

        return squin.broadcast.measure(ancillas)

    @squin.kernel
    def physical_rotation_noisy(data: Register):
        for idx in range(0, num_data, d):
            squin.rz(angle_rad, data[idx])
            if p1 > 0.0:
                squin.depolarize(p=p1, qubit=data[idx])

    return inizialize_qubits_noisy, syndrome_round_noisy, physical_rotation_noisy
```

```python
def make_distance_n_surface_2rounds(d:int):
    z_supports, x_supports = build_stabilizer_supports(d)
    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    inizialize_qubits, syndrome_round, phys_rot = make_helpers(d, 0)

    @squin.kernel
    def circuit() -> Measurement:
        data     = squin.qalloc(num_data)
        ancillas = squin.qalloc(num_ancillas)

        m0 = inizialize_qubits(data, ancillas)

        m1 = syndrome_round(data, ancillas)
        m2 = syndrome_round(data, ancillas)

        for i in range(num_ancillas):
            squin.set_detector([m1[i], m2[i]], [])

        return m2

    return circuit

```

The provided circuits are for a default rotation angle of 0.01*pi. To simulate different rotation angles, you can use the following function to compute the physical rotation angle needed to achieve a logical rotation of angle `logical_angle_in_pi` on `num_physical_rotations` physical rotations.

Also make sure to check out the comments in `circuits/star_d=3.stim` to get some hints about the circuit structure.

**Goal:** Learn how to load and run pre-built circuits using Tsim. 


```python
# Using import from_file

import stim, re
import tsim as ts
import matplotlib.pyplot as plt

def logical_to_physical_angle(logical_angle_in_pi: float, num_physical_rotations: int) -> float:
    assert (
        num_physical_rotations % 2 == 1 and num_physical_rotations > 0
    ), "k must be a positive odd integer"
    sign = -1 if (num_physical_rotations + 1) % 4 == 0 else 1
    logical_angle_in_rad = logical_angle_in_pi * np.pi
    x = np.tan(logical_angle_in_rad / 2) ** (1 / num_physical_rotations)
    theta_phys = 2 * np.arctan(x)
    return float(sign * theta_phys / np.pi)


def load_star_circuit(d: int, logical_angle_in_pi: float) -> ts.Circuit:
    descr = ""
    rz_found = 0
    new_phys = logical_to_physical_angle(logical_angle_in_pi, d)
    with open(f"assets/star_circuits/star_d={d}.stim", "r") as inp:
        lines = inp.readlines()
        for idx, l in enumerate(lines):
            if l.startswith("R_Z"): # R_Z(-0.01) 6 2 3 4
                if rz_found == 0:
                    l = re.sub(r"R_Z\([^)]+\)", f"R_Z({new_phys})", l)
                else:
                    l = re.sub(r"R_Z\([^)]+\)", f"R_Z({logical_angle_in_pi})", l)

                lines[idx] = l
                rz_found += 1

    descr = """\n""".join(lines)
    return ts.Circuit(descr)
```

```python
import numpy as np
def make_star_rotation(d, logical_angle, MUTE_DETECTORS=False):
    z_supports, x_supports = build_stabilizer_supports(d)
    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    phys_angle = logical_to_physical_angle(logical_angle/np.pi, d)*np.pi
    inizialize_qubits, syndrome_round, physical_rotation = make_helpers(d, phys_angle)

    @squin.kernel
    def circuit():
        data     = squin.qalloc(num_data)
        ancillas = squin.qalloc(num_ancillas)

        inizialize_qubits(data, ancillas)

        m0 = syndrome_round(data, ancillas)
        m1 = syndrome_round(data, ancillas)
        if(not MUTE_DETECTORS):
            for idx in range(num_ancillas):
                squin.set_detector([m0[idx], m1[idx]], [])

        physical_rotation(data)
        for _ in range(d):
            syndrome_round(data, ancillas)

        # logical unrotation via CNOT chain — stop before last row to avoid OOB
        for idx in range(0, num_data - d, d):
            squin.cx(data[idx], data[idx + d])

        squin.rz(-logical_angle, data[num_data - d])

        # reverse CNOT chain — same bounds
        for idx in range(num_data - 2 * d, -1, -d):
            squin.cx(data[idx], data[idx + d])

        # measure logical X: H then measure first row as IList
        for q in range(d):
            squin.h(data[q])
        m_logical_x = squin.broadcast.measure(data[:d])

        if(not MUTE_DETECTORS):
            squin.set_observable(m_logical_x, 0)
        return m_logical_x

    return circuit
```

```python
import numpy as np


def make_star_rotation_noisy(
    d: int,
    logical_angle_rad: float,
    p1: float = 1e-3,
    p2: float = 1e-3,
    INSERT_DETECTORS: bool = True,
    noisy_final_readout: bool = False,
):
    """
    Noisy version of make_star_rotation.

    logical_angle_rad:
        Logical angle in radians, i.e. what you want on the x-axis of the plot.

    p1:
        Depolarizing noise after 1-qubit gates.

    p2:
        Depolarizing noise after 2-qubit gates.

    noisy_final_readout:
        False = the final unrotation/logical readout part remains noiseless.
        True  = also adds noise to the final part.
    """

    z_supports, x_supports = build_stabilizer_supports(d)

    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    # logical_to_physical_angle expects the logical angle in units of pi.
    logical_angle_in_pi = logical_angle_rad / np.pi

    # Returns physical angle in units of pi.
    phys_angle_in_pi = logical_to_physical_angle(logical_angle_in_pi, d)

    # squin.rz(...) expects radians.
    phys_angle_rad = phys_angle_in_pi * np.pi

    inizialize_qubits, syndrome_round, physical_rotation = make_helpers_noisy(
        d=d,
        angle_rad=phys_angle_rad,
        p1=p1,
        p2=p2,
    )

    @squin.kernel
    def circuit():
        data     = squin.qalloc(num_data)
        ancillas = squin.qalloc(num_ancillas)

        inizialize_qubits(data, ancillas)


        m0 = syndrome_round(data, ancillas)
        m1 = syndrome_round(data, ancillas)

        if INSERT_DETECTORS:
            for idx in range(num_ancillas):
                squin.set_detector([m0[idx], m1[idx]], [])

        physical_rotation(data)
        for _ in range(d):
            syndrome_round(data, ancillas)

        # Logical unrotation via CNOT chain.
        # By default this part remains noiseless, as in the official STAR files.
        for idx in range(0, num_data - d, d):
            squin.cx(data[idx], data[idx + d])
            if noisy_final_readout and p2 > 0.0:
                squin.depolarize2(p2, data[idx], data[idx + d])

        squin.rz(-logical_angle_rad, data[num_data - d])
        if noisy_final_readout and p1 > 0.0:
            squin.depolarize(p=p1, qubit=data[num_data - d])

        for idx in range(num_data - 2 * d, -1, -d):
            squin.cx(data[idx], data[idx + d])
            if noisy_final_readout and p2 > 0.0:
                squin.depolarize2(p2, data[idx], data[idx + d])

        # Measure logical X: H then measure first row.
        for q in range(d):
            squin.h(data[q])
            if noisy_final_readout and p1 > 0.0:
                squin.depolarize(p=p1, qubit=data[q])

        m_logical_x = squin.broadcast.measure(data[:d])

        if INSERT_DETECTORS:
            squin.set_observable(m_logical_x, 0)

        return m_logical_x

    return circuit
```

```python
def make_star_rotation_noisy_v2(
    d: int,
    logical_angle_rad: float,
    p1: float = 1e-3,
    p2: float = 1e-3,
    INSERT_DETECTORS: bool = False,
    noisy_final_readout: bool = False,
):
    """
    Version closer to the structure of the official STAR files.

    Main difference from the previous version:
    - prepares |+>
    - runs a full syndrome round before the rotation
    - then applies the physical STAR rotation
    - then runs two rounds for gadget postselection
    - then runs d extra rounds
    - then clean final round + logical unrotation + X_L measurement
    """

    z_supports, x_supports = build_stabilizer_supports(d)

    num_data     = d**2
    num_z        = len(z_supports)
    num_x        = len(x_supports)
    num_ancillas = num_z + num_x

    logical_angle_in_pi = logical_angle_rad / np.pi
    phys_angle_in_pi = logical_to_physical_angle(logical_angle_in_pi, d)
    phys_angle_rad = phys_angle_in_pi * np.pi

    inizialize_qubits, syndrome_round, physical_rotation = make_helpers_noisy(
        d=d,
        angle_rad=phys_angle_rad,
        p1=p1,
        p2=p2,
    )

    # Clean helpers for the noiseless final part.
    _, syndrome_round_clean, _ = make_helpers(
        d=d,
        angle=0.0,
    )

    @squin.kernel
    def circuit():
        data     = squin.qalloc(num_data)
        ancillas = squin.qalloc(num_ancillas)

        # 1. |+> preparation + initial Z stabilizer measurement
        m_init_z = inizialize_qubits(data, ancillas)

        # 2. First full syndrome round BEFORE rotation
        m_pre = syndrome_round(data, ancillas)

        if INSERT_DETECTORS:
            # Z detector: compare Z from the first full round with the initial Z measurement
            for i in range(num_z):
                squin.set_detector([m_pre[i], m_init_z[i]], [])

            # X detector: in the first round X stabilizers are compared against 0
            for i in range(num_x):
                squin.set_detector([m_pre[num_z + i]], [])

        # 3. STAR physical rotation

        # 4. First round after rotation
        m_after_0 = syndrome_round(data, ancillas)

        if INSERT_DETECTORS:
            # As in the official files: after rotation, Z detectors are singletons,
            # X detectors are compared with the previous round.
            for i in range(num_z):
                squin.set_detector([m_after_0[i]], [])

            for i in range(num_x):
                squin.set_detector([m_after_0[num_z + i], m_pre[num_z + i]], [])

        # 5. Second round after rotation
        m_after_1 = syndrome_round(data, ancillas)

        if INSERT_DETECTORS:
            for i in range(num_ancillas):
                squin.set_detector([m_after_1[i], m_after_0[i]], [])

        # At this point the first 3*(d^2 - 1) detectors are those to be postselected.

        previous = m_after_1

        physical_rotation(data)
        # 6. d extra stabilizer measurement rounds
        for _ in range(d):
            current = syndrome_round(data, ancillas)

            if INSERT_DETECTORS:
                for i in range(num_ancillas):
                    squin.set_detector([current[i], previous[i]], [])

            previous = current

        # 7. Clean final round, as in the official files after DISABLE_ERROR
        final_round = syndrome_round_clean(data, ancillas)

        if INSERT_DETECTORS:
            for i in range(num_ancillas):
                squin.set_detector([final_round[i], previous[i]], [])

        # 8. Perfect logical unrotation on the first column
        for idx in range(0, num_data - d, d):
            squin.cx(data[idx], data[idx + d])
            if noisy_final_readout and p2 > 0.0:
                squin.depolarize2(p2, data[idx], data[idx + d])

        squin.rz(-logical_angle_rad, data[num_data - d])
        if noisy_final_readout and p1 > 0.0:
            squin.depolarize(p=p1, qubit=data[num_data - d])

        for idx in range(num_data - 2 * d, -1, -d):
            squin.cx(data[idx], data[idx + d])
            if noisy_final_readout and p2 > 0.0:
                squin.depolarize2(p2, data[idx], data[idx + d])

        # 9. X_L measurement on the first row, as MPP X0*X1*...*X(d-1)
        for q in range(d):
            squin.h(data[q])
            if noisy_final_readout and p1 > 0.0:
                squin.depolarize(p=p1, qubit=data[q])

        m_logical_x = squin.broadcast.measure(data[:d])

        if INSERT_DETECTORS:
            squin.set_observable(m_logical_x, 0)

        return m_logical_x

    return circuit
```

```python
import pymatching

def run_one_star_point(
    d: int,
    theta_rad: float,
    shots: int = 1000,
    p1: float = 1e-3,
    p2: float = 1e-3,
    seed: int = 0,
    postselect: bool = True,
    ):
    """
    Simulates a single data point of the plot.

    d:
        code distance.

    theta_rad:
        logical angle in radians, i.e. the x-axis of the plot.

    shots:
        number of shots.

    p1, p2:
        depolarizing noise after 1-qubit and 2-qubit gates.

    postselect:
        if True, keeps only shots where the decoder predicts no logical error
        on the first n_post detectors (the STAR gadget rounds).
    """

    kernel = make_star_rotation_noisy_v2(
        d=d,
        logical_angle_rad=theta_rad,
        p1=p1,
        p2=p2,
        INSERT_DETECTORS=True,
        noisy_final_readout=False,
    )
    
    # kernel = make_star_rotation(
    #     d=d,
    #     logical_angle=theta_rad,
    #     MUTE_DETECTORS=False,
    # )

    circuit = compile_circuit_from_squin_kernel(kernel, seed=seed)
    sampler = circuit.compile_detector_sampler(seed=seed)

    detector_error_model = circuit.detector_error_model()
    matcher = pymatching.Matching.from_detector_error_model(detector_error_model)

    detectors, observables = sampler.sample(
        shots=shots,
        separate_observables=True,
    )

    detectors = np.asarray(detectors)
    observables = np.asarray(observables)

    if detectors.ndim == 1:
        detectors = detectors.reshape(shots, -1)

    if observables.ndim == 1:
        observables = observables.reshape(shots, -1)

    n_post = 3 * (d * d - 1)
    n_post = min(n_post, detectors.shape[1])

    if postselect and n_post > 0:
        # Zero out detectors beyond the first n_post so the matcher only
        # sees the STAR gadget rounds; any predicted correction means reject.
        det_masked = detectors.copy()
        det_masked[:, n_post:] = 0
        predictions = np.asarray(matcher.decode_batch(det_masked))
        if predictions.ndim == 1:
            predictions = predictions.reshape(-1, 1)
        accepted_mask = np.all(predictions == 0, axis=1)
    else:
        accepted_mask = np.ones(shots, dtype=bool)

    print(f"  detectors={detectors.shape[1]}, postselected_detectors={n_post}")

    accepted_shots = int(np.sum(accepted_mask))
    acceptance_rate = accepted_shots / shots

    if accepted_shots == 0:
        logical_errors_observed = 0

        # Conditional error undefined when there are no accepted shots
        conditional_logical_error_rate = np.nan
        conditional_stderr = np.nan
    else:
        logical_bits = observables[accepted_mask, 0].astype(np.uint8)
        logical_errors_observed = int(np.sum(logical_bits))

        # Old metric: errors / accepted shots
        conditional_logical_error_rate = logical_errors_observed / accepted_shots
        conditional_stderr = float(
            np.sqrt(
                conditional_logical_error_rate
                * (1.0 - conditional_logical_error_rate)
                / accepted_shots
            )
        )

    # New main metric: errors / total shots
    logical_error_rate = logical_errors_observed / shots

    stderr = float(
        np.sqrt(
            logical_error_rate * (1.0 - logical_error_rate) / shots
        )
    )

    return {
        "d": d,
        "theta_rad": theta_rad,
        "theta_in_pi": theta_rad / np.pi,
        "shots": shots,
        "accepted_shots": accepted_shots,
        "acceptance_rate": acceptance_rate,

        "logical_errors_observed": logical_errors_observed,

        # Main metric for the plot:
        # logical errors / total shots
        "logical_error_rate": logical_error_rate,
        "logical_error_rate_over_d": logical_error_rate / d if not np.isnan(logical_error_rate) else np.nan,
        "stderr": stderr,
        "stderr_over_d": stderr / d if not np.isnan(stderr) else np.nan,

        # Secondary metric useful for debugging:
        # logical errors / accepted shots
        "conditional_logical_error_rate": conditional_logical_error_rate,
        "conditional_logical_error_rate_over_d": (
            conditional_logical_error_rate / d
            if not np.isnan(conditional_logical_error_rate)
            else np.nan
        ),
        "conditional_stderr": conditional_stderr,
        "conditional_stderr_over_d": (
            conditional_stderr / d
            if not np.isnan(conditional_stderr)
            else np.nan
        ),

        "p1": p1,
        "p2": p2,
    }


def run_one_star_point_adaptive(
    d: int,
    theta_rad: float,
    target_accepted: int = 1000,
    max_shots: int = 10_000_000,
    batch_shots: int = 100_000,
    p1: float = 1e-3,
    p2: float = 1e-3,
    seed: int = 0,
    postselect: bool = True,
):
    """
    Simulates a point until it reaches target_accepted accepted shots,
    or until it exceeds max_shots.

    Useful for d=5 and d=7, where postselection is very rare.
    """

    kernel = make_star_rotation_noisy_v2(
        d=d,
        logical_angle_rad=theta_rad,
        p1=p1,
        p2=p2,
        INSERT_DETECTORS=True,
        noisy_final_readout=False,
    )

    circuit = compile_circuit_from_squin_kernel(kernel, seed=seed)
    sampler = circuit.compile_detector_sampler(seed=seed)

    dem = circuit.detector_error_model(decompose_errors=False)
    matching = pymatching.Matching.from_detector_error_model(dem)

    total_shots = 0
    total_accepted = 0
    total_logical_errors = 0

    n_post = 3 * (d * d - 1)

    while total_shots < max_shots and total_accepted < target_accepted:
        current_batch = min(batch_shots, max_shots - total_shots)

        detectors, observables = sampler.sample(
            shots=current_batch,
            separate_observables=True,
        )

        detectors = np.asarray(detectors)
        observables = np.asarray(observables)

        if detectors.ndim == 1:
            detectors = detectors.reshape(current_batch, -1)

        if observables.ndim == 1:
            observables = observables.reshape(current_batch, -1)

        pred = matching.decode_batch(detectors)
        if pred.ndim == 1:
            pred = pred.reshape(-1, 1)

        obs_corrected = observables ^ pred

        accepted = current_batch
        logical_bits = obs_corrected[:, 0].astype(np.uint8)
        logical_errors = int(np.sum(logical_bits))

        total_shots += current_batch
        total_accepted += accepted
        total_logical_errors += logical_errors

        print(
            f"  batch done: total_shots={total_shots}, "
            f"accepted={total_accepted}, errors={total_logical_errors}"
        )

    acceptance_rate = total_accepted / total_shots if total_shots > 0 else np.nan

    if total_shots == 0:
        logical_error_rate = np.nan
        stderr = np.nan
    else:
        logical_error_rate = total_logical_errors / total_shots
        stderr = np.sqrt(
            logical_error_rate * (1.0 - logical_error_rate) / total_shots
        )

    return {
        "d": d,
        "theta_rad": theta_rad,
        "theta_in_pi": theta_rad / np.pi,
        "shots": total_shots,
        "accepted_shots": total_accepted,
        "acceptance_rate": acceptance_rate,
        "logical_errors_observed": total_logical_errors,
        "logical_error_rate": logical_error_rate,
        "logical_error_rate_over_d": logical_error_rate / d if not np.isnan(logical_error_rate) else np.nan,
        "stderr": stderr,
        "stderr_over_d": stderr / d if not np.isnan(stderr) else np.nan,
        "p1": p1,
        "p2": p2,
        "postselected_detectors": n_post,
    }
```

```python
def run_star_sweep(
    distances=(3, 5, 7),
    theta_values_rad=None,
    shots: int = 1000,
    p1: float = 1e-3,
    p2: float = 1e-3,
    postselect: bool = True,
):
    """
    Runs the sweep over distances and angles.
    """

    if theta_values_rad is None:
        theta_values_rad = np.geomspace(1e-4, 1.6, 8)

    rows = []

    for d in distances:
        for i, theta_rad in enumerate(theta_values_rad):
            print(f"Running d={d}, theta={theta_rad:.4g} rad, shots={shots}...")

            row = run_one_star_point(
                d=d,
                theta_rad=theta_rad,
                shots=shots,
                p1=p1,
                p2=p2,
                seed=1234 + 100 * d + i,
                postselect=postselect,
            )

            print(
                f"  accepted={row['accepted_shots']}/{shots} "
                f"acc={row['acceptance_rate']:.4f} "
                f"pL={row['logical_error_rate']:.4e} "
                f"pL/d={row['logical_error_rate_over_d']:.4e}"
            )

            rows.append(row)

    df = pd.DataFrame(rows)
    df.to_csv("custom_star_tsim_results.csv", index=False)

    return df

def plot_star_results(df, filename="custom_star_tsim_plot.png"):
    qubits = {
        3: 17,
        5: 49,
        7: 97,
    }

    markers = {
        3: "o",
        5: "s",
        7: "^",
    }

    fig, ax = plt.subplots(figsize=(7.2, 4.8))

    for d, group in df.groupby("d"):
        group = group.sort_values("theta_rad")

        # Values of 0 cannot be plotted on a log scale.
        group_plot = group[group["logical_error_rate_over_d"] > 0]

        if len(group_plot) == 0:
            print(f"Warning: for d={d} all values are 0 or NaN. Increase shots.")
            continue

        ax.errorbar(
            group_plot["theta_rad"],
            group_plot["logical_error_rate_over_d"],
            yerr=group_plot["stderr_over_d"],
            marker=markers.get(d, "o"),
            linewidth=2,
            markersize=6,
            capsize=3,
            label=f"$d = {d}$ ({qubits.get(d, d**2)} qubits)",
        )

    ax.set_xscale("log")
    ax.set_yscale("log")

    ax.set_xlabel(r"Rotation angle $\theta$ (rad)")
    ax.set_ylabel("Logical error rate / distance")

    ax.grid(True, which="both", alpha=0.3)
    ax.legend()

    fig.tight_layout()
    fig.savefig(filename, dpi=300)
    plt.show()

    print(f"Plot saved to: {filename}")
```

## Team STAR Part 3 — Teleport a non-Clifford rotation into a logical qubit
Now assume a noiseless setting but where non-Clifford gates cannot be applied directly to the main logical qubit. 
Construct a protocol that uses an ancillary logical qubit to teleport a small-angle rotation into the main qubit while assuming the STAR transversal architecture.
Important: you will need to use postselection to filter results because Tsim (unlike PyQrack) does not support feed-forwarded operations.

![rus.png](./assets/rus.png)

Figure taken from *"Partially Fault-Tolerant Quantum Computing Architecture with Error-Corrected Clifford Gates and Space-Time Efficient Analog Rotations"* by *Akahoshi et al.*

**Goal:** Show how STAR enables indirect implementation of small-angle non-Clifford logical rotations, and analyze the costs of doing so. 


```python
import numpy as np

def make_star_teleport_rotation_noisy(d, logical_angle, p1, p2, INSERT_DETECTORS=True):
    """
    Teleports Rz_L(logical_angle) onto the main logical qubit using a
    STAR-prepared ancilla logical qubit.

    Circuit (Akahoshi et al.):

        |psi>_L  --[+]-- M_Z_L
        |m0>_L   --[*]-- X_L  -->  Rz_L(+-theta)|psi>_L

    |m0>_L = Rz_L(theta)|+>_L is prepared via the STAR transversal protocol.
    The logical CNOT is implemented transversally (physical CX per data qubit).
    M_Z_L is the logical Z measurement of the main qubit.

    Since Tsim does not support feed-forward, we postselect on M_Z_L = 0,
    which selects the branch where no X_L correction is needed.

    p1: depolarizing noise probability after 1-qubit gates.
    p2: depolarizing noise probability after 2-qubit gates.

    Noise is applied to: qubit initialization, all syndrome rounds, transversal CNOT.
    The final logical unrotation and readout are noiseless (as in make_star_rotation_noisy).

    Observables:
      obs[:,0]  -- M_Z_L of main qubit  (postselect on 0)
      obs[:,1]  -- M_X_L of ancilla     (the output rotation signal)
    """
    z_supports, x_supports = build_stabilizer_supports(d)
    num_data = d**2
    num_z    = len(z_supports)
    num_x    = len(x_supports)
    num_ancs = num_z + num_x

    # logical_to_physical_angle returns pi units; make_helpers_noisy expects radians
    phys_angle_rad = logical_to_physical_angle(logical_angle/np.pi, d) * np.pi

    # Noisy helpers for STAR ancilla: init prepares |+>_L, phys_rot = transversal Rz + noise
    init_q, syn_noisy, phys_rot = make_helpers_noisy(d, phys_angle_rad, p1, p2)
    # Noisy init for main qubit (|+>_L + noisy Z stabilizer measurement, no rotation)
    init_m, _, _ = make_helpers_noisy(d, 0.0, p1, p2)
    # Noiseless helpers for final readout
    _, syn_clean, _ = make_helpers(d, 0)

    @squin.kernel
    def circuit():
        data_m = squin.qalloc(num_data)
        stab_m = squin.qalloc(num_ancs)
        data_s = squin.qalloc(num_data)
        stab_s = squin.qalloc(num_ancs)

        # Prepare main in |+>_L (noisy)
        init_m(data_m, stab_m)

        # Prepare STAR ancilla Rz_L(theta)|+>_L (noisy)
        init_q(data_s, stab_s)
        phys_rot(data_s)

        # Two stabilizer rounds on each qubit (noisy)
        m0m = syn_noisy(data_m, stab_m)
        m0s = syn_noisy(data_s, stab_s)
        m1m = syn_noisy(data_m, stab_m)
        m1s = syn_noisy(data_s, stab_s)

        if INSERT_DETECTORS:
            for i in range(num_ancs):
                squin.set_detector([m0m[i], m1m[i]], [])
                squin.set_detector([m0s[i], m1s[i]], [])

        # Transversal logical CNOT: control=ancilla, target=main (noisy)
        for i in range(num_data):
            squin.cx(data_s[i], data_m[i])
            if p2 > 0.0:
                squin.depolarize2(p2, data_s[i], data_m[i])

        # Post-CNOT stabilizer rounds (noisy)
        m2m = syn_noisy(data_m, stab_m)
        m2s = syn_noisy(data_s, stab_s)

        if INSERT_DETECTORS:
            for i in range(num_ancs):
                squin.set_detector([m2m[i], m1m[i]], [])
                squin.set_detector([m2s[i], m1s[i]], [])

        # ── Noiseless Z_L measurement of main qubit ────────────────────────
        for col in range(0, num_data - d, d):
            squin.cx(data_m[col], data_m[col + d])
        m_z_main = squin.broadcast.measure(data_m[num_data - d : num_data - d + 1])

        # ── Noiseless logical unrotation + X_L readout of STAR ancilla ─────
        for col in range(0, num_data - d, d):
            squin.cx(data_s[col], data_s[col + d])
        squin.rz(-logical_angle, data_s[num_data - d])
        for col in range(num_data - 2 * d, -1, -d):
            squin.cx(data_s[col], data_s[col + d])
        for q in range(d):
            squin.h(data_s[q])
        m_x_star = squin.broadcast.measure(data_s[:d])

        if INSERT_DETECTORS:
            squin.set_observable(m_z_main, 0)   # postselect on 0 -> correct branch
            squin.set_observable(m_x_star, 1)   # logical X output of rotated ancilla

        return m_x_star

    return circuit
```

```python

# ── Demo: sweep logical angle and compare teleported vs. direct STAR ──────────
import matplotlib.pyplot as plt

shots     = 2000
angles_pi = np.linspace(0.01, 1.0, 12)   # logical angle in units of pi
p1, p2 = 1e-4, 1e-4

fig, ax = plt.subplots(figsize=(7, 4.5))

for d in [3, 5]:
    print("START d=", d)
    errs_teleport = []
    errs_direct   = []

    for a in angles_pi:
        print("Angle ", a)
        # ── teleported rotation ──────────────────────────────────────────
        circ_t    = squin_to_circuit(make_star_teleport_rotation_noisy(d=d, logical_angle=a, p1=p1, p2=p2))
        sampler_t = circ_t.compile_detector_sampler(seed=42)
        dets_t, obs_t = sampler_t.sample(shots, separate_observables=True)
        dets_t = np.asarray(dets_t);  obs_t = np.asarray(obs_t)
        clean_t = (dets_t == 0).all(axis=1) & (obs_t[:, 0] == 0)
        errs_teleport.append(float(obs_t[clean_t, 1].mean()) if clean_t.sum() > 0 else np.nan)

        # ── direct STAR rotation (reference) ────────────────────────────
        circ_d    = squin_to_circuit(make_star_rotation_noisy(d=d, logical_angle_rad=a * np.pi, p1=p1, p2=p2, INSERT_DETECTORS=True))
        sampler_d = circ_d.compile_detector_sampler(seed=42)
        dets_d, obs_d = sampler_d.sample(shots, separate_observables=True)
        dets_d = np.asarray(dets_d);  obs_d = np.asarray(obs_d)
        clean_d = (dets_d == 0).all(axis=1)
        errs_direct.append(float(obs_d[clean_d, 0].mean()) if clean_d.sum() > 0 else np.nan)

    color = "tab:blue" if d == 3 else "tab:orange"
    ax.plot(angles_pi * np.pi, errs_teleport, "o-",  color=color, label=f"d={d} teleport")
    ax.plot(angles_pi * np.pi, errs_direct,   "x--", color=color, label=f"d={d} direct STAR", alpha=0.6)

ax.set_xlabel(r"Logical rotation angle $\theta_L$ (rad)")
ax.set_ylabel(r"$\langle X_L \rangle$ (logical error proxy)")
ax.set_title("STAR Rotation Teleportation vs Direct STAR (noisy, p1=p2=1e-3)")
ax.loglog()
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```


## Team STAR Part 4: Apply STAR in a larger algorithm
Construct a 4-qubit QFT circuit in which all non-Clifford gates must be teleported in rather than applied directly.

**Goal:** Demonstrate how teleported non-Clifford gates can be integrated into a larger logical circuit, and evaluate the scalability of the approach.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ─── 4-qubit QFT circuit structure ───────────────────────────────────────────
#
# Standard 4-qubit QFT (before bit-reversal swaps):
#
#  q0: ─H─CR2(q0,q1)─CR3(q0,q2)─CR4(q0,q3)─ ...
#  q1: ─────────────H─CR2(q1,q2)─CR3(q1,q3)─ ...
#  q2: ────────────────────────H─CR2(q2,q3)─ ...
#  q3: ─────────────────────────────────────H
#
#  CR_k applies controlled-phase CP(φ) with φ = 2π/2^k = π/2^{k-1}:
#    k=2: CP(π/2) = CS gate              ← Clifford, applied directly
#    k=3: CP(π/4), decomposes to Rz(π/8) ← non-Clifford → STAR teleportation
#    k=4: CP(π/8), decomposes to Rz(π/16)← non-Clifford → STAR teleportation
#
# CP(φ) decomposition (up to global phase):
#   CP(φ) = Rz_ctrl(+φ/2) · CNOT(ctrl→tgt) · Rz_tgt(-φ/2) · CNOT(ctrl→tgt)
#
# Non-Clifford Rz gates produced by decomposing all non-Clifford CP gates:
#   CR3(q0,q2): Rz(+π/8) on q0,   Rz(-π/8) on q2
#   CR3(q1,q3): Rz(+π/8) on q1,   Rz(-π/8) on q3
#   CR4(q0,q3): Rz(+π/16) on q0,  Rz(-π/16) on q3
#
# Total: 6 STAR teleportation calls per QFT execution.
# Clifford gates (H, CS, CNOT, SWAP) are applied directly without teleportation.

NC_GATES = [
    # (gate, qubit_label, role,    angle_rad)
    ("CR3", "q0", "ctrl", +np.pi / 8),
    ("CR3", "q2", "tgt",  -np.pi / 8),
    ("CR3", "q1", "ctrl", +np.pi / 8),
    ("CR3", "q3", "tgt",  -np.pi / 8),
    ("CR4", "q0", "ctrl", +np.pi / 16),
    ("CR4", "q3", "tgt",  -np.pi / 16),
]


def simulate_qft4_star_overhead(d, p1=1e-3, p2=1e-3, shots=2000, seed=0):
    """
    Estimate the STAR teleportation overhead for each non-Clifford Rz gate in
    the 4-qubit QFT.  Each gate's acceptance rate is the fraction of shots
    where all syndrome detectors are 0 AND M_Z_L = 0 (correct postselection
    branch).  The joint success probability is the product over all 6 gates.
    """
    per_gate = []

    for gi, (gate_name, qubit, role, angle_rad) in enumerate(NC_GATES):
        circ = squin_to_circuit(
            make_star_teleport_rotation_noisy(
                d=d,
                logical_angle=abs(angle_rad),   # radians; sign handled by unrotation
                p1=p1,
                p2=p2,
                INSERT_DETECTORS=True,
            )
        )
        sampler = circ.compile_detector_sampler(seed=seed + gi * 7)
        dets, obs = sampler.sample(shots, separate_observables=True)
        dets = np.asarray(dets)
        obs  = np.asarray(obs)

        clean = (dets == 0).all(axis=1) & (obs[:, 0] == 0)
        acc   = float(clean.sum()) / shots

        per_gate.append({
            "gate": gate_name, "qubit": qubit, "role": role,
            "angle_rad": angle_rad, "acceptance_rate": acc,
        })

    acc_arr = np.array([g["acceptance_rate"] for g in per_gate])
    joint   = float(np.prod(acc_arr))

    return {
        "d": d, "p1": p1, "p2": p2,
        "per_gate": per_gate,
        "joint_success_prob":      joint,
        "expected_trials_per_qft": 1.0 / joint if joint > 0 else np.inf,
    }


# ─── Sweep: distance × noise level ───────────────────────────────────────────
distances    = [3, 5]
noise_levels = [5e-4, 1e-3, 3e-3, 1e-2]
shots        = 3000

sweep_rows = []
for d in distances:
    for p in noise_levels:
        print(f"  d={d}, p={p:.1e} ...", end=" ", flush=True)
        res = simulate_qft4_star_overhead(d=d, p1=p, p2=p, shots=shots, seed=42)
        sweep_rows.append({
            "d":                  d,
            "p":                  p,
            "joint_success_prob": res["joint_success_prob"],
            "expected_trials":    res["expected_trials_per_qft"],
            "mean_acceptance":    float(np.mean([g["acceptance_rate"] for g in res["per_gate"]])),
        })
        print(f"P_joint={res['joint_success_prob']:.4f}  E[trials]={res['expected_trials_per_qft']:.1f}")

df4 = pd.DataFrame(sweep_rows)

# ─── Plots ────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

# Left: expected trials to successfully run one QFT vs noise (log-log)
ls = {3: "o-", 5: "s--"}
for d in distances:
    sub = df4[df4["d"] == d]
    axes[0].loglog(sub["p"], sub["expected_trials"], ls[d], label=f"d={d}")
axes[0].set_xlabel("Physical noise p  (p1 = p2 = p)")
axes[0].set_ylabel("Expected trials for one successful QFT")
axes[0].set_title("4-qubit QFT: total STAR overhead vs noise")
axes[0].legend()
axes[0].grid(True, which="both", alpha=0.3)

# Right: per-gate acceptance rate bar chart at p = 1e-3
p_ref = 1e-3
gate_labels = [f"{g['gate']}\n({g['qubit']},{g['role']})" for g in NC_GATES]
x = np.arange(len(NC_GATES))
for k, d in enumerate(distances):
    res  = simulate_qft4_star_overhead(d=d, p1=p_ref, p2=p_ref, shots=shots, seed=0)
    accs = [g["acceptance_rate"] for g in res["per_gate"]]
    axes[1].bar(x + (k - 0.5) * 0.35, accs, width=0.35, label=f"d={d}", alpha=0.85)
axes[1].set_xticks(x)
axes[1].set_xticklabels(gate_labels, fontsize=8)
axes[1].set_ylim(0, 1)
axes[1].set_ylabel("Per-gate acceptance rate")
axes[1].set_title(f"Per-gate STAR acceptance  (p = {p_ref:.0e})")
axes[1].legend()
axes[1].grid(axis="y", alpha=0.3)

fig.tight_layout()
plt.savefig("qft4_star_overhead.png", dpi=150)
plt.show()

print("\n── Summary ───────────────────────────────────────────────────────────")
print(df4.to_string(index=False, float_format="{:.4g}".format))
print()
print("'expected_trials': number of full QFT attempts needed until all 6")
print("non-Clifford STAR teleportations succeed (all post-selections pass).")
```
