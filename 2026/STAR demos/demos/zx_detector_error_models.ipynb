```python
from tsim import Circuit
```

```python
c = Circuit("""
    X_ERROR(0.0) 0 # Dummy error that ensures 1-indexing like in https://arxiv.org/pdf/2407.13826
    R 0 4
    X_ERROR(0.1) 1 2 3
    TICK
    CNOT 1 0 2 4 2 0 3 4
    X_ERROR(0.1) 0 4
    TICK
    M 0 4
    X_ERROR(0.1)  1 2 3
    TICK
    M 1 2 3
    DETECTOR rec[-5]
    DETECTOR rec[-4]
    DETECTOR rec[-5] rec[-3] rec[-2]
    DETECTOR rec[-4] rec[-2] rec[-1]
""")
```

```python
c.diagram(height=200)
```

```python
c.diagram("pyzx", scale=35);
```

```python
c.diagram("pyzx-dets", scale_horizontally=1.5)#, transform_error_basis=False);
```
