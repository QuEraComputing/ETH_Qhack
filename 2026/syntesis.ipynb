```python
# ============================================================
# Part 3 — Postselection dropout without feed-forward correction
# ============================================================

import os, sys, importlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bloqade.pyqrack import StackMemorySimulator

sys.path.insert(0, os.path.dirname(os.path.abspath("syntesis_1_2.py")))

import gate_syntesis; importlib.reload(gate_syntesis)
from gate_syntesis import make_postselected_dropout_kernel, t_count_from_sequence

N_VALUES      = globals().get("N_VALUES",      range(51))
PART2_EPSILON = globals().get("PART2_EPSILON", 1e-10)
RESULTS_DIR   = globals().get("RESULTS_DIR",   "results")

if "part2_circuits" not in globals() or "part2_sequences" not in globals():
    import postselection; importlib.reload(postselection)
    part2_circuits, part2_sequences = postselection.ensure_part2_results(N_VALUES, PART2_EPSILON)

# --- Exact dropout probabilities ---
rows = []
for n in N_VALUES:
    t = t_count_from_sequence(part2_sequences[n])
    rows.append({"n": n, "logical_T_count": t,
                 "postselection_success_probability_exact": 2**-t,
                 "dropout_probability_exact": 1 - 2**-t})
df_exact = pd.DataFrame(rows)
display(df_exact)

# --- Shot-based estimates (only for small T-count) ---
DROPOUT_SIM_SHOTS     = 200
MAX_SIMULATED_T_COUNT = 2

sim_rows = []
for n in N_VALUES:
    t = t_count_from_sequence(part2_sequences[n])
    if t > MAX_SIMULATED_T_COUNT:
        continue
    if t == 0:
        sim_rows.append({"n": n, "logical_T_count": t, "shots": 0, "dropout_rate_estimated": 0.0})
        continue
    kernel   = make_postselected_dropout_kernel(part2_sequences[n])
    task     = StackMemorySimulator(min_qubits=1 + t).task(kernel)
    dropouts = sum(int(task.run()) for _ in range(DROPOUT_SIM_SHOTS))
    sim_rows.append({"n": n, "logical_T_count": t,
                     "shots": DROPOUT_SIM_SHOTS, "dropout_rate_estimated": dropouts / DROPOUT_SIM_SHOTS})
df_sim = pd.DataFrame(sim_rows)
display(df_sim)

# --- Plot ---
os.makedirs(RESULTS_DIR, exist_ok=True)
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(df_exact["n"], df_exact["dropout_probability_exact"], marker="o", label="Exact")
if not df_sim.empty:
    est = df_sim[df_sim["shots"] > 0]
    ax.scatter(est["n"], est["dropout_rate_estimated"], marker="x", label="Shot estimate")
ax.set(xlabel="n", ylabel="Dropout probability",
       title="Part 3: postselection dropout without feed-forward",
       ylim=(-0.05, 1.05))
ax.set_xticks(df_exact["n"])
ax.legend(); ax.grid(True)
fig.tight_layout()
fig.savefig(os.path.join(RESULTS_DIR, "part3_postselection_dropout_probability.png"), dpi=300, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# Part 4 - Simulate one synthesized Rz sequence on Steane [[7,1,3]]
# ============================================================

import os
import sys
import importlib

import numpy as np
from bloqade import tsim
from bloqade.pyqrack import StackMemorySimulator

sys.path.insert(0, os.getcwd())

import gate_syntesis_helpers as gate_syntesis_helpers
import gate_syntesis

importlib.reload(gate_syntesis_helpers)
importlib.reload(gate_syntesis)

from gate_syntesis import (
    make_part4_logical_fidelity_kernel_reuse_magic,
    make_part4_tsim_diagram_kernel,
    part4_logical_fidelity_qubit_count,
)
from gate_syntesis_helpers import t_count_from_sequence
import gate_syntesis_helpers as gh

PART4_N = 2

part4_sequence = tuple(part2_sequences[PART4_N])
part4_t_count  = t_count_from_sequence(part4_sequence)

print(f"Part 4 logical simulation for n={PART4_N}")
print("sequence = " + " ".join(part4_sequence).upper())
print(f"T-count  = {part4_t_count}")

# ── Circuit diagram (Stim/tsim — no dynamic feed-forward shown) ──────────────
diagram_kernel = make_part4_tsim_diagram_kernel(part4_sequence)
display(tsim.Circuit(diagram_kernel).diagram(width=1200))

# ── True simulation with feed-forward correction ──────────────────────────────
kernel               = make_part4_logical_fidelity_kernel_reuse_magic(part4_sequence)
part4_physical_qubits = part4_logical_fidelity_qubit_count(part4_sequence)

sim        = StackMemorySimulator(min_qubits=part4_physical_qubits)
data_block = sim.run(kernel)
rho_data   = sim.reduced_density_matrix(data_block)

print(f"\nphysical qubits simulated = {part4_physical_qubits}")
print(f"returned data block size  = {len(data_block)}")
print(f"reduced density matrix shape = {rho_data.shape}")
print(f"trace(rho_data) = {np.trace(rho_data)}")

# ── Fidelity against the target logical state ─────────────────────────────────

def _steane_basis_single(logical_bit):
    psi = np.zeros(2**7, dtype=complex)
    for a in (0, 1):
        for b in (0, 1):
            for c in (0, 1):
                bits = [a, b, a^b, c, a^c, b^c, a^b^c]
                if logical_bit == 1:
                    bits = [x ^ 1 for x in bits]
                idx = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[idx] += 1.0 / np.sqrt(8)
    return psi

zero_L = _steane_basis_single(0)
one_L  = _steane_basis_single(1)

theta    = np.pi / (2 ** PART4_N)
psi_tgt  = (np.exp(-0.5j * theta) * zero_L + np.exp(0.5j * theta) * one_L) / np.sqrt(2)
fidelity = float(np.clip(np.real(np.vdot(psi_tgt, rho_data @ psi_tgt)), 0.0, 1.0))
print(f"\nLogical state fidelity vs target Rz_L(π/2^{PART4_N})|+_L⟩ = {fidelity:.10f}")
```

```python
# ============================================================
# Part 4 — Rz_L(π/2^n) con Steane_apply_logical_gate_sequence_reuse_magic
#           epsilon = 1e-10, magic block singolo riusabile
# ============================================================

import importlib
import numpy as np
import pandas as pd
from bloqade import squin
from bloqade.pyqrack import StackMemorySimulator

import gate_syntesis_helpers as gh
import gate_syntesis
importlib.reload(gh)
importlib.reload(gate_syntesis)

from gate_syntesis import (
    Steane_zero_logical_graph,
    Steane_H_logical,
    Steane_apply_logical_gate_sequence,
    Steane_apply_logical_gate_sequence_reuse_magic,
)
from qiskit.synthesis import gridsynth_rz

EPSILON_10 = 1e-10
N_REUSE    = range(6)   # cambia il range a piacere


# ── sequenza Part 2 ──────────────────────────────────────────────────────────

def _seq(n, eps=EPSILON_10):
    if n == 0: return ("s", "s")
    if n == 1: return ("s",)
    if n == 2: return ("t",)
    return gh.gate_sequence_from_circuit(gridsynth_rz(np.pi / 2**n, epsilon=eps))


# ── kernel factory ────────────────────────────────────────────────────────────

def build_reuse_kernel(sequence):
    """
    Kernel che esegue, a livello logico Steane [[7,1,3]]:
      1. logical |0⟩   via Steane_zero_logical_graph()
      2. logical |+⟩   via Steane_H_logical()
      3. magic block   squin.qalloc(7) — ripreparato come |A_L⟩ per ogni T
      4. Steane_apply_logical_gate_sequence_reuse_magic(data, magic, seq)
    """
    sequence = tuple(sequence)
    t_count  = gh.t_count_from_sequence(sequence)

    if t_count == 0:
        # Nessun gate T: il magic block non serve, usa la versione senza reuse
        @squin.kernel
        def _k():
            data_block = Steane_zero_logical_graph()   # passo 1 – logical |0⟩
            Steane_H_logical(data_block)               # passo 2 – logical |+⟩
            Steane_apply_logical_gate_sequence(data_block, [], sequence)
            return data_block
        return _k, 7

    @squin.kernel
    def _k():
        # passo 1 – prepara logical |0⟩ tramite il circuito grafo-stato Steane [[7,1,3]]
        data_block = Steane_zero_logical_graph()

        # passo 2 – applica H logico → logical |+⟩  (= H|0_L⟩)
        Steane_H_logical(data_block)

        # passo 3 – alloca UN blocco magic (7 fisici) riusabile:
        #           prima di ogni gate T verrà ripreparato come |A_L⟩ = T_L|+_L⟩
        #           e resettato a |0⟩ dopo l'iniezione, pronto per il prossimo T
        magic_block = squin.qalloc(7)

        # passo 4 – esegui la sequenza Clifford+T a livello logico con magic-state reuse
        Steane_apply_logical_gate_sequence_reuse_magic(data_block, magic_block, sequence)

        return data_block   # soli i 7 qubit del data block

    return _k, 14  # 7 data + 7 magic


# ── Steane basis states per fidelity ─────────────────────────────────────────

def _steane_basis(logical_bit):
    psi = np.zeros(2**7, dtype=complex)
    for a in (0, 1):
        for b in (0, 1):
            for c in (0, 1):
                bits = [a, b, a^b, c, a^c, b^c, a^b^c]
                if logical_bit == 1:
                    bits = [x ^ 1 for x in bits]
                idx = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[idx] += 1.0 / np.sqrt(8)
    return psi

_ZERO_L = _steane_basis(0)
_ONE_L  = _steane_basis(1)

def _target_psi_L(theta):
    """Stato logico target dopo H_L poi Rz_L(theta): (e^{-iθ/2}|0_L⟩ + e^{iθ/2}|1_L⟩)/√2"""
    return (np.exp(-0.5j * theta) * _ZERO_L + np.exp(0.5j * theta) * _ONE_L) / np.sqrt(2)


# ── simulazione ───────────────────────────────────────────────────────────────

rows = []

for n in N_REUSE:
    theta    = np.pi / (2**n)
    sequence = _seq(n, EPSILON_10)
    t_count  = gh.t_count_from_sequence(sequence)

    seq_str  = ' '.join(sequence).upper()
    print(f"n={n}  T-count={t_count:3d}  seq={seq_str[:55]}{'…' if len(seq_str)>55 else ''}")

    kernel, n_qubits = build_reuse_kernel(sequence)
    sim              = StackMemorySimulator(min_qubits=n_qubits)
    data_block       = sim.run(kernel)
    rho              = sim.reduced_density_matrix(data_block)

    psi_tgt  = _target_psi_L(theta)
    fidelity = float(np.clip(np.real(np.vdot(psi_tgt, rho @ psi_tgt)), 0.0, 1.0))

    rows.append({
        "n":           n,
        "sequence":    seq_str,
        "T_count":     t_count,
        "phys_qubits": n_qubits,
        "fidelity":    fidelity,
        "infidelity":  max(0.0, 1.0 - fidelity),
    })
    print(f"       → fidelity = {fidelity:.10f}")

df_reuse = pd.DataFrame(rows)
display(df_reuse[["n", "T_count", "phys_qubits", "fidelity", "infidelity"]])
```

```python
# ============================================================
# Part 4 - Logical-state fidelity on Steane [[7,1,3]]
# ============================================================

import os
import sys
import importlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from bloqade.pyqrack import StackMemorySimulator

sys.path.insert(0, os.getcwd())

import gate_syntesis_helpers as gh
import gate_syntesis

importlib.reload(gh)
importlib.reload(gate_syntesis)

from gate_syntesis import (
    make_part4_logical_fidelity_kernel_reuse_magic,
    part4_logical_fidelity_qubit_count,
)


PART4_FIDELITY_N_VALUES = range(0, 8)
PART4_FIDELITY_SHOTS = 30


def get_part4_sequence(n: int):
    if "part2_sequences" in globals() and n in part2_sequences:
        return tuple(part2_sequences[n])

    if "part2_rz_sequence" in globals():
        epsilon = globals().get("PART2_EPSILON", globals().get("EPSILON", 1e-4))
        return tuple(part2_rz_sequence(n, epsilon))

    raise NameError(
        "part2_sequences is not defined. Run the Part 2 sequence cell before this cell."
    )


def steane_basis_state(logical_bit: int):
    # Return |0_L> or |1_L> in the basis order used by reduced_density_matrix.
    psi = np.zeros(2**7, dtype=complex)

    for a in (0, 1):
        for b in (0, 1):
            for c in (0, 1):
                bits = [a, b, a ^ b, c, a ^ c, b ^ c, a ^ b ^ c]

                if logical_bit == 1:
                    bits = [bit ^ 1 for bit in bits]

                # PyQrack reduced_density_matrix uses big-endian matrix order
                # for the qubit list [q0, ..., q6].
                index = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[index] += 1 / np.sqrt(8)

    return psi


ZERO_L = steane_basis_state(0)
ONE_L = steane_basis_state(1)


def target_logical_state_after_H_then_Rz(theta: float):
    return (
        np.exp(-0.5j * theta) * ZERO_L
        + np.exp(0.5j * theta) * ONE_L
    ) / np.sqrt(2)


def pure_state_fidelity_from_rho(rho, psi):
    value = np.vdot(psi, rho @ psi)
    return float(np.clip(np.real(value), 0.0, 1.0))


def ideal_logical_state_fidelity_from_sequence(theta: float, sequence):
    # Ideal logical-level prediction for the same synthesized Clifford+T sequence.
    psi0 = np.array([1, 0], dtype=complex)
    psi_after_h = gh.H @ psi0
    psi_target = gh.Rz(theta) @ psi_after_h
    psi_sequence = gh.unitary_from_gate_sequence(sequence) @ psi_after_h
    fidelity = abs(np.vdot(psi_target, psi_sequence)) ** 2
    return float(np.clip(np.real(fidelity), 0.0, 1.0))


rows = []

for n in PART4_FIDELITY_N_VALUES:
    theta = np.pi / (2**n)
    sequence = get_part4_sequence(n)
    t_count = gh.t_count_from_sequence(sequence)

    print(f"Simulating Part 4 n={n}, T-count={t_count}, shots={PART4_FIDELITY_SHOTS}")

    kernel = make_part4_logical_fidelity_kernel_reuse_magic(sequence)
    physical_qubits_simulated = part4_logical_fidelity_qubit_count(sequence)
    psi_target_logical = target_logical_state_after_H_then_Rz(theta)

    fidelities = []
    for _ in range(PART4_FIDELITY_SHOTS):
        sim = StackMemorySimulator(min_qubits=physical_qubits_simulated)
        data_block = sim.run(kernel)
        rho_data = sim.reduced_density_matrix(data_block)
        fidelities.append(pure_state_fidelity_from_rho(rho_data, psi_target_logical))

    mean_fidelity = float(np.mean(fidelities))
    std_fidelity = float(np.std(fidelities))
    ideal_fidelity = ideal_logical_state_fidelity_from_sequence(theta, sequence)

    rows.append(
        {
            "n": n,
            "target": f"H_L then Rz_L(pi/2^{n})",
            "Rz_sequence": " ".join(sequence).upper(),
            "T_or_Tdg_count": t_count,
            "physical_qubits_simulated": physical_qubits_simulated,
            "shots": PART4_FIDELITY_SHOTS,
            "steane_state_fidelity_mean": mean_fidelity,
            "steane_state_fidelity_std": std_fidelity,
            "steane_state_infidelity_mean": max(0.0, 1.0 - mean_fidelity),
            "ideal_logical_state_fidelity": ideal_fidelity,
            "ideal_logical_state_infidelity": max(0.0, 1.0 - ideal_fidelity),
        }
    )


df_part4_logical_fidelity = pd.DataFrame(rows)
#display(df_part4_logical_fidelity)

fig, ax = plt.subplots(figsize=(7, 4))
ax.errorbar(
    df_part4_logical_fidelity["n"],
    df_part4_logical_fidelity["steane_state_fidelity_mean"],
    yerr=df_part4_logical_fidelity["steane_state_fidelity_std"],
    marker="o",
    capsize=3,
    label="Full Steane simulation",
)
ax.plot(
    df_part4_logical_fidelity["n"],
    df_part4_logical_fidelity["ideal_logical_state_fidelity"],
    marker="s",
    linestyle="--",
    label="Ideal logical prediction",
)
ax.set_xlabel("n")
ax.set_ylabel("Logical state fidelity")
ax.set_title("Part 4: logical fidelity after H_L then Rz_L")
ax.set_xticks(list(PART4_FIDELITY_N_VALUES))
ax.set_ylim(0.999999, 1.000001)
ax.ticklabel_format(axis="y", style="plain", useOffset=False)
ax.yaxis.set_major_formatter(plt.FormatStrFormatter("%.7f"))
ax.legend()
ax.grid(True)
fig.tight_layout()
plt.show()

positive_steane_infidelity = df_part4_logical_fidelity["steane_state_infidelity_mean"].replace(0.0, np.nan)
positive_ideal_infidelity = df_part4_logical_fidelity["ideal_logical_state_infidelity"].replace(0.0, np.nan)

fig, ax = plt.subplots(figsize=(7, 4))
ax.semilogy(
    df_part4_logical_fidelity["n"],
    positive_steane_infidelity,
    marker="o",
    label="Full Steane simulation",
)
ax.semilogy(
    df_part4_logical_fidelity["n"],
    positive_ideal_infidelity,
    marker="s",
    linestyle="--",
    label="Ideal logical prediction",
)
ax.set_xlabel("n")
ax.set_ylabel("Logical state infidelity 1 - F")
ax.set_title("Part 4: logical state infidelity")
ax.set_xticks(list(PART4_FIDELITY_N_VALUES))
ax.legend()
ax.grid(True)
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from qiskit.synthesis import gridsynth_rz

import gate_syntesis_helpers as gh


EPSILON = 1e-6
N_VALUES = range(50)


def matrix_gate_distance(U, V):
    value = 1 - abs(np.trace(U.conj().T @ V)) / 2
    return np.sqrt(max(0.0, float(np.real(value))))


def safe_gridsynth_rz_sequence(theta: float, epsilon: float):
    target = gh.Rz(theta)

    # For coarse tolerances, the identity circuit can already be a valid
    # approximation. This also avoids a known gridsynth_rz panic at epsilon=1e-1.
    if matrix_gate_distance(target, gh.I) <= epsilon:
        return ()

    # If the Rust gridsynth backend panics for a coarse epsilon, retry with a
    # stricter tolerance. A stricter approximation is still valid for the
    # original epsilon.
    candidate_epsilons = [epsilon, 1e-3, 1e-4, 1e-6, 1e-10]
    tried = set()

    for eps_try in candidate_epsilons:
        if eps_try in tried or eps_try > epsilon:
            continue

        tried.add(eps_try)

        try:
            qiskit_circuit = gridsynth_rz(theta, epsilon=eps_try)
        except BaseException as exc:
            if type(exc)._name_ != "PanicException":
                raise
            continue

        sequence = gh.gate_sequence_from_circuit(qiskit_circuit)
        distance = matrix_gate_distance(target, gh.unitary_from_gate_sequence(sequence))

        if distance <= epsilon + 1e-15:
            return sequence

    raise RuntimeError(
        f"Could not synthesize Rz({theta}) within epsilon={epsilon} without gridsynth_rz failing."
    )


def part2_rz_sequence(n: int, epsilon: float = EPSILON):
    # Exact cases.
    if n == 0:
        return ("s", "s")   # Rz(pi) = Z = S^2 up to global phase

    if n == 1:
        return ("s",)       # Rz(pi/2) = S up to global phase

    if n == 2:
        return ("t",)       # Rz(pi/4) = T up to global phase

    # Approximate cases.
    theta = np.pi / (2 ** n)
    return safe_gridsynth_rz_sequence(theta, epsilon)


def unitary_fidelity(U, V):
    value = abs(np.trace(U.conj().T @ V)) / 2

    # The physical fidelity is in [0, 1]. Tiny excursions above 1 come only
    # from floating-point roundoff and make the plot look misleading.
    return float(np.clip(np.real(value), 0.0, 1.0))


def safe_gate_distance(U, V):
    fidelity = unitary_fidelity(U, V)
    value = max(0.0, 1.0 - fidelity)
    return np.sqrt(value)


def sparse_n_ticks(values, step=5):
    values = list(values)
    if len(values) <= 15:
        return values
    return values[::step]


def disable_y_offset(ax):
    formatter = ScalarFormatter(useOffset=False)
    formatter.set_scientific(False)
    ax.yaxis.set_major_formatter(formatter)


rows = []

for n in N_VALUES:
    theta = np.pi / (2 ** n)

    sequence = part2_rz_sequence(n, EPSILON)

    U_target = gh.Rz(theta)
    V_circuit = gh.unitary_from_gate_sequence(sequence)

    fidelity = unitary_fidelity(U_target, V_circuit)
    distance = safe_gate_distance(U_target, V_circuit)
    infidelity = max(0.0, 1.0 - fidelity)

    rows.append(
        {
            "n": n,
            "target": f"Rz(pi/2^{n})",
            "theta": theta,
            "sequence": " ".join(sequence).upper(),
            "gate_count": len(sequence),
            "T_or_Tdg_count": gh.t_count_from_sequence(sequence),
            "unitary_fidelity": fidelity,
            "challenge_distance": distance,
            "infidelity": infidelity,
        }
    )


df_p2_fidelity = pd.DataFrame(rows)
#display(df_p2_fidelity)


plt.figure(figsize=(7, 4))
plt.plot(df_p2_fidelity["n"], df_p2_fidelity["unitary_fidelity"], marker="o")
plt.xlabel("n")
plt.ylabel("Unitary fidelity")
plt.title("Part 2: unitary fidelity, zoomed near 1")
plt.xticks(sparse_n_ticks(N_VALUES))
plt.ylim(0.999999999999, 1.000000000001)
disable_y_offset(plt.gca())
plt.grid(True)
plt.tight_layout()
plt.show()


positive_infidelity = df_p2_fidelity["infidelity"].replace(0.0, np.nan)

plt.figure(figsize=(7, 4))
plt.semilogy(df_p2_fidelity["n"], positive_infidelity, marker="o")
plt.xlabel("n")
plt.ylabel("Infidelity 1 - F")
plt.title("Part 2: synthesis error, zeros omitted on log scale")
plt.xticks(sparse_n_ticks(N_VALUES))
plt.grid(True)
plt.tight_layout()
plt.show()


positive_distance = df_p2_fidelity["challenge_distance"].replace(0.0, np.nan)

plt.figure(figsize=(7, 4))
plt.semilogy(df_p2_fidelity["n"], positive_distance, marker="o")
plt.xlabel("n")
plt.ylabel("Challenge distance d(U,V)")
plt.title("Part 2: challenge distance, zeros omitted on log scale")
plt.xticks(sparse_n_ticks(N_VALUES))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from qiskit.synthesis import gridsynth_rz
from qiskit.quantum_info import Statevector, state_fidelity

import gate_syntesis_helpers as gh


EPSILON = 1e-4
N_VALUES = range(20)


def matrix_gate_distance(U, V):
    value = 1 - abs(np.trace(U.conj().T @ V)) / 2
    return np.sqrt(max(0.0, float(np.real(value))))


def safe_gridsynth_rz_sequence(theta: float, epsilon: float):
    target = gh.Rz(theta)

    # For coarse tolerances, the identity circuit can already be a valid
    # approximation. This also avoids a known gridsynth_rz panic at epsilon=1e-1.
    if matrix_gate_distance(target, gh.I) <= epsilon:
        return ()

    # If the Rust gridsynth backend panics for a coarse epsilon, retry with a
    # stricter tolerance. A stricter approximation is still valid for the
    # original epsilon.
    candidate_epsilons = [epsilon, 1e-3, 1e-4, 1e-6, 1e-10]
    tried = set()

    for eps_try in candidate_epsilons:
        if eps_try in tried or eps_try > epsilon:
            continue

        tried.add(eps_try)

        try:
            qiskit_circuit = gridsynth_rz(theta, epsilon=eps_try)
        except BaseException as exc:
            if type(exc)._name_ != "PanicException":
                raise
            continue

        sequence = gh.gate_sequence_from_circuit(qiskit_circuit)
        distance = matrix_gate_distance(target, gh.unitary_from_gate_sequence(sequence))

        if distance <= epsilon + 1e-15:
            return sequence

    raise RuntimeError(
        f"Could not synthesize Rz({theta}) within epsilon={epsilon} without gridsynth_rz failing."
    )


def part2_rz_sequence(n: int, epsilon: float = EPSILON):
    """
    Return the Clifford+T sequence used in Part 2 for Rz(pi/2^n).
    The first three cases are exact.
    """
    if n == 0:
        return ("s", "s")   # Rz(pi) = Z = S^2 up to global phase

    if n == 1:
        return ("s",)       # Rz(pi/2) = S up to global phase

    if n == 2:
        return ("t",)       # Rz(pi/4) = T up to global phase

    theta = np.pi / (2 ** n)
    return safe_gridsynth_rz_sequence(theta, epsilon)


psi0 = np.array([1, 0], dtype=complex)
psi_after_H = gh.H @ psi0

rows = []

for n in N_VALUES:
    theta = np.pi / (2 ** n)

    rz_sequence = part2_rz_sequence(n, EPSILON)

    U_target = gh.Rz(theta)
    V_synthesized = gh.unitary_from_gate_sequence(rz_sequence)

    psi_target = U_target @ psi_after_H
    psi_synthesized = V_synthesized @ psi_after_H

    fidelity = state_fidelity(
        Statevector(psi_target),
        Statevector(psi_synthesized),
    )

    rows.append(
        {
            "n": n,
            "target": f"H then Rz(pi/2^{n})",
            "part2_sequence_after_H": " ".join(("h",) + rz_sequence).upper(),
            "Rz_sequence": " ".join(rz_sequence).upper(),
            "T_or_Tdg_count": gh.t_count_from_sequence(rz_sequence),
            "state_fidelity": fidelity,
            "state_infidelity": 1 - fidelity,
        }
    )


df_H_then_Rz_state_fidelity = pd.DataFrame(rows)
#display(df_H_then_Rz_state_fidelity)


plt.figure(figsize=(7, 4))
plt.plot(
    df_H_then_Rz_state_fidelity["n"],
    df_H_then_Rz_state_fidelity["state_fidelity"],
    marker="o",
)
plt.xlabel("n")
plt.ylabel("State fidelity")
plt.title("Part 2 check: |0> -> H -> Rz(pi/2^n)")
plt.xticks(list(N_VALUES))
plt.grid(True)
plt.ticklabel_format(axis="y", style="plain", useOffset=False)
plt.show()


plt.figure(figsize=(7, 4))
plt.semilogy(
    df_H_then_Rz_state_fidelity["n"],
    df_H_then_Rz_state_fidelity["state_infidelity"].clip(lower=1e-18),
    marker="o",
)
plt.xlabel("n")
plt.ylabel("State infidelity 1 - F")
plt.title("Part 2 check: state infidelity after H then Rz(pi/2^n)")
plt.xticks(list(N_VALUES))
plt.grid(True)
plt.show()
```

```python
# ============================================================
# Part 4 — Logical simulation with strategic depolarizing noise
# ============================================================
#
# Noise is added selectively at the dominant error sources:
#   • squin.depolarize2(p) after every CX in the T-injection CNOT_L gadget
#   • squin.depolarize2(p) after every CX in the magic-state encoding circuit
#   • squin.depolarize(p/10) after the physical T gate in magic-state prep
# All transversal Clifford gates (H_L, S_L, logical encoding) are noiseless.
# Pattern mirrors the tutorial §Noise Models (closure-factory approach).

import importlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bloqade import squin
from bloqade.pyqrack import StackMemorySimulator
from qiskit.synthesis import gridsynth_rz

import gate_syntesis_helpers as gh
import gate_syntesis
importlib.reload(gh)
importlib.reload(gate_syntesis)

from gate_syntesis import (
    Steane_zero_logical_graph,
    Steane_H_logical, Steane_S_logical, Steane_Sdg_logical,
    Steane_T_logical_reset, Steane_apply_logical_gate_sequence,
)
from gate_syntesis_helpers import Steane_measure_logical_Z_weight3, gate_sequence_from_circuit


# ── Factory: noisy CNOT_L (depolarize2 after every CX) ───────────────────
def make_noisy_cnot_logical(p2q: float):
    @squin.kernel
    def noisy_cnot(ctrl, tgt):
        for i in range(7):
            squin.cx(ctrl[i], tgt[i])
            squin.depolarize2(p=p2q, control=ctrl[i], target=tgt[i])
        return ctrl + tgt
    return noisy_cnot


# ── Factory: noisy magic-state encoding (noise after T + each CX) ────────
def make_noisy_magic_prep(p2q: float, p1q_T: float):
    @squin.kernel
    def noisy_magic(q):
        squin.h(q[0]); squin.h(q[1]); squin.h(q[3]); squin.h(q[2])
        squin.t(q[2]); squin.depolarize(p=p1q_T, qubit=q[2])
        squin.cx(q[0], q[6]); squin.depolarize2(p=p2q, control=q[0], target=q[6])
        squin.cx(q[1], q[5]); squin.depolarize2(p=p2q, control=q[1], target=q[5])
        squin.cx(q[2], q[4]); squin.depolarize2(p=p2q, control=q[2], target=q[4])
        squin.cx(q[1], q[6]); squin.depolarize2(p=p2q, control=q[1], target=q[6])
        squin.cx(q[3], q[4]); squin.depolarize2(p=p2q, control=q[3], target=q[4])
        squin.cx(q[2], q[5]); squin.depolarize2(p=p2q, control=q[2], target=q[5])
        squin.cx(q[1], q[2]); squin.depolarize2(p=p2q, control=q[1], target=q[2])
        squin.cx(q[3], q[5]); squin.depolarize2(p=p2q, control=q[3], target=q[5])
        squin.cx(q[0], q[4]); squin.depolarize2(p=p2q, control=q[0], target=q[4])
        squin.cx(q[0], q[2]); squin.depolarize2(p=p2q, control=q[0], target=q[2])
        squin.cx(q[3], q[6]); squin.depolarize2(p=p2q, control=q[3], target=q[6])
        return q
    return noisy_magic


# ── Factory: noisy T injection gadget (uses noisy CNOT_L, keeps feed-forward) ──
def make_noisy_t_injection(noisy_cnot_L):
    @squin.kernel
    def noisy_t_inj(data, magic):
        noisy_cnot_L(data, magic)
        m = Steane_measure_logical_Z_weight3(magic)
        if m:
            Steane_S_logical(data)
        Steane_T_logical_reset(magic)
        return data
    return noisy_t_inj


def make_noisy_tdg_injection(noisy_cnot_L):
    @squin.kernel
    def noisy_tdg_inj(data, magic):
        noisy_cnot_L(data, magic)
        m = Steane_measure_logical_Z_weight3(magic)
        if m:
            Steane_S_logical(data)
        Steane_Sdg_logical(data)
        Steane_T_logical_reset(magic)
        return data
    return noisy_tdg_inj


# ── Factory: sequence applier (captures noisy building blocks as closures) ──
def make_noisy_apply_seq(noisy_magic, noisy_t_inj, noisy_tdg_inj):
    @squin.kernel
    def noisy_apply(data, magic, seq):
        for gate in seq:
            if gate == "h":
                Steane_H_logical(data)
            elif gate == "s":
                Steane_S_logical(data)
            elif gate == "sdg":
                Steane_Sdg_logical(data)
            elif gate == "t":
                noisy_magic(magic)
                noisy_t_inj(data, magic)
            elif gate == "tdg":
                noisy_magic(magic)
                noisy_tdg_inj(data, magic)
        return data
    return noisy_apply


# ── Factory: full Part 4 noisy kernel ────────────────────────────────────
def make_part4_noisy_kernel(sequence, p_noise: float):
    """H_L then Rz_L sequence on Steane [[7,1,3]] with strategic circuit-level noise.
    Noise fires only on T-injection gadgets; Clifford parts are noiseless."""
    sequence = tuple(sequence)
    t_count  = gh.t_count_from_sequence(sequence)
    p2q      = p_noise
    p1q_T    = p_noise / 10

    if t_count == 0:
        @squin.kernel
        def _k_clifford():
            data = Steane_zero_logical_graph()
            Steane_H_logical(data)
            Steane_apply_logical_gate_sequence(data, [], sequence)
            return data
        return _k_clifford, 7

    noisy_cnot_L  = make_noisy_cnot_logical(p2q)
    noisy_magic   = make_noisy_magic_prep(p2q, p1q_T)
    noisy_t_inj   = make_noisy_t_injection(noisy_cnot_L)
    noisy_tdg_inj = make_noisy_tdg_injection(noisy_cnot_L)
    noisy_apply   = make_noisy_apply_seq(noisy_magic, noisy_t_inj, noisy_tdg_inj)

    @squin.kernel
    def _k():
        data  = Steane_zero_logical_graph()
        Steane_H_logical(data)
        magic = squin.qalloc(7)
        noisy_apply(data, magic, sequence)
        return data
    return _k, 14


# ── Steane logical basis and target state ─────────────────────────────────
def _steane_basis_n(b: int):
    psi = np.zeros(2**7, dtype=complex)
    for a in (0, 1):
        for b_ in (0, 1):
            for c in (0, 1):
                bits = [a, b_, a^b_, c, a^c, b_^c, a^b_^c]
                if b == 1:
                    bits = [x ^ 1 for x in bits]
                idx = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[idx] += 1.0 / np.sqrt(8)
    return psi

_ZL = _steane_basis_n(0)
_OL = _steane_basis_n(1)

def _target(n: int):
    theta = np.pi / 2**n
    return (np.exp(-0.5j * theta) * _ZL + np.exp(0.5j * theta) * _OL) / np.sqrt(2)

def _fid(rho, psi):
    return float(np.clip(np.real(np.vdot(psi, rho @ psi)), 0.0, 1.0))


# ── Get Clifford+T sequence for n ─────────────────────────────────────────
def _get_seq(n: int):
    if "part2_sequences" in globals() and n in part2_sequences:
        return tuple(part2_sequences[n])
    if n == 0: return ("h", "s", "s", "h")
    if n == 1: return ("s",)
    if n == 2: return ("t",)
    return tuple(gate_sequence_from_circuit(gridsynth_rz(np.pi / 2**n, epsilon=1e-4)))


# ── Simulate ──────────────────────────────────────────────────────────────
NOISY_SHOTS    = 3
NOISY_N_VALUES = [0, 1, 2, 3]
P_FIXED        = 0.01    # noise level for the n-sweep
N_FIXED        = 2       # n value for the noise sweep
P_SWEEP        = [0.0, 0.005, 0.01, 0.03, 0.05]


def simulate_noisy(n: int, p: float):
    seq              = _get_seq(n)
    kernel, n_qubits = make_part4_noisy_kernel(seq, p)
    psi_t            = _target(n)
    fids = []
    for _ in range(NOISY_SHOTS):
        sim  = StackMemorySimulator(min_qubits=n_qubits)
        data = sim.run(kernel)
        rho  = sim.reduced_density_matrix(data)
        fids.append(_fid(rho, psi_t))
    return float(np.mean(fids)), float(np.std(fids))


print("=== Part 4 noisy: n-sweep (p = {:.3f}) ===".format(P_FIXED))
rows_n = []
for n in NOISY_N_VALUES:
    mi, si = simulate_noisy(n, 0.0)
    mn, sn = simulate_noisy(n, P_FIXED)
    print(f"  n={n}: F_ideal={mi:.5f}  F_noisy={mn:.5f}")
    rows_n.append({"n": n, "F_ideal": mi, "F_ideal_std": si, "F_noisy": mn, "F_noisy_std": sn})

print(f"\n=== Noise sweep (n = {N_FIXED}) ===")
rows_p = []
for p in P_SWEEP:
    m, s = simulate_noisy(N_FIXED, p)
    print(f"  p={p:.3f}: F = {m:.5f} \u00b1 {s:.5f}")
    rows_p.append({"p_noise": p, "F_mean": m, "F_std": s})

df_n = pd.DataFrame(rows_n)
df_p = pd.DataFrame(rows_p)

# ── Plots ─────────────────────────────────────────────────────────────────
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(13, 4))

ax0.errorbar(df_n["n"], df_n["F_ideal"], yerr=df_n["F_ideal_std"],
             marker="s", ls="--", capsize=4, label="noiseless  (p = 0)")
ax0.errorbar(df_n["n"], df_n["F_noisy"], yerr=df_n["F_noisy_std"],
             marker="o", capsize=4, label=f"strategic noise  (p\u2082q = {P_FIXED})")
ax0.set_xlabel("n  (target rotation Rz(\u03c0/2\u207f))")
ax0.set_ylabel("Logical state fidelity")
ax0.set_title("Steane [[7,1,3]]: fidelity vs n")
ax0.set_xticks(NOISY_N_VALUES)
ax0.legend(); ax0.grid(True)

ax1.errorbar(df_p["p_noise"], df_p["F_mean"], yerr=df_p["F_std"],
             marker="o", capsize=4, color="C1")
ax1.set_xlabel("Depolarizing probability  p\u2082q  (per CX gate)")
ax1.set_ylabel("Logical state fidelity")
ax1.set_title(f"Steane [[7,1,3]]: fidelity vs noise  (n = {N_FIXED})")
ax1.grid(True)

fig.suptitle(
    "Part 4 \u2014 Strategic circuit-level noise\n"
    "Noise on: CX in CNOT_L, CX in magic-state encoding, T in magic-state prep",
    fontsize=10,
)
fig.tight_layout()
plt.savefig(f"{RESULTS_DIR}/part4_strategic_noise.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"\nNote: n=0,1 use only Clifford gates \u2014 unaffected by T-injection noise (F \u2248 1).")
```

```python
# ============================================================
# Part 4 — Logical noise via GeminiOne/TwoZoneNoiseModel
#
# Mirrors the tutorial §Noise Models workflow (emit → transform → load):
#   1. squin kernel  →  emit_circuit  →  cirq.Circuit
#   2. noise.transform_circuit(model=GeminiOne/TwoZoneNoiseModel())
#   3. load_circuit  →  noisy squin kernel
#
# Target circuit: |+_L> = H_L|0_L> (7-qubit Clifford encoding).
# Using the encoding circuit avoids mid-circuit feed-forward (not
# supported by Cirq emit) while still exercising atom moves, site
# pairing, and single-qubit gate noise from the Gemini architecture.
# Note: this cell is simulation-heavy; reduce GEMINI_SHOTS if needed.
# ============================================================

import importlib
import numpy as np
import matplotlib.pyplot as plt
from bloqade import squin
from bloqade.cirq_utils import emit_circuit, load_circuit, noise
from bloqade.pyqrack import StackMemorySimulator

import gate_syntesis_helpers as gh
import gate_syntesis
importlib.reload(gh)
importlib.reload(gate_syntesis)

from gate_syntesis import Steane_zero_logical_graph, Steane_H_logical


# ── Target circuit: prepare |+_L> = H_L|0_L> (7 qubits, all Clifford) ───
@squin.kernel
def steane_plus_L():
    q = Steane_zero_logical_graph()   # logical |0_L> via graph-state encoding
    Steane_H_logical(q)               # logical H  ->  |+_L>
    return q


# ── Step 1: emit squin kernel -> cirq.Circuit ─────────────────────────────
print("Emitting Steane |+_L> encoding circuit to Cirq ...")
cirq_circuit = emit_circuit(steane_plus_L, ignore_returns=True)
print(f"  Original circuit: {len(cirq_circuit)} moment(s), "
      f"{len(cirq_circuit.all_qubits())} qubit(s)")


# ── Step 2: transform with Gemini noise models ────────────────────────────
print("Applying GeminiOneZoneNoiseModel ...")
noisy_cirq_one = noise.transform_circuit(
    cirq_circuit, model=noise.GeminiOneZoneNoiseModel()
)
print(f"  One-zone noisy circuit: {len(noisy_cirq_one)} moment(s)")

print("Applying GeminiTwoZoneNoiseModel ...")
noisy_cirq_two = noise.transform_circuit(
    cirq_circuit, model=noise.GeminiTwoZoneNoiseModel()
)
print(f"  Two-zone noisy circuit: {len(noisy_cirq_two)} moment(s)")


# ── Step 3: load noisy cirq circuits back into squin ─────────────────────
noisy_squin_one = load_circuit(
    noisy_cirq_one, kernel_name="steane_plusL_one_zone",
    register_as_argument=True, return_register=True,
)
noisy_squin_two = load_circuit(
    noisy_cirq_two, kernel_name="steane_plusL_two_zone",
    register_as_argument=True, return_register=True,
)


# ── Target state: |+_L> = (|0_L> + |1_L>) / sqrt(2) ─────────────────────
def _steane_basis_g(b: int):
    psi = np.zeros(2**7, dtype=complex)
    for a in (0, 1):
        for b_ in (0, 1):
            for c in (0, 1):
                bits = [a, b_, a^b_, c, a^c, b_^c, a^b_^c]
                if b == 1:
                    bits = [x ^ 1 for x in bits]
                idx = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[idx] += 1.0 / np.sqrt(8)
    return psi

_ZL_G    = _steane_basis_g(0)
_OL_G    = _steane_basis_g(1)
psi_plus = (_ZL_G + _OL_G) / np.sqrt(2)

def _fid_g(rho, psi):
    return float(np.clip(np.real(np.vdot(psi, rho @ psi)), 0.0, 1.0))


# ── Build wrapper kernels (one per noise variant) ─────────────────────────
def _make_runner(noisy_k):
    @squin.kernel
    def _runner():
        reg = squin.qalloc(7)
        reg = noisy_k(reg)
        return reg
    return _runner

runner_one = _make_runner(noisy_squin_one)
runner_two = _make_runner(noisy_squin_two)


# ── Simulate ──────────────────────────────────────────────────────────────
GEMINI_SHOTS = 3  # reduce if too slow

def run_shots(kernel, label):
    fids = []
    for s in range(GEMINI_SHOTS):
        print(f"  [{label}] shot {s+1}/{GEMINI_SHOTS} ...", end=" ", flush=True)
        sim  = StackMemorySimulator(min_qubits=7)
        data = sim.run(kernel)
        rho  = sim.reduced_density_matrix(data)
        f    = _fid_g(rho, psi_plus)
        fids.append(f)
        print(f"F = {f:.5f}")
    return fids

print("\nNoiseless baseline ...")
fids_ideal = run_shots(steane_plus_L, "Ideal")

print("\nGeminiOneZoneNoiseModel ...")
fids_one = run_shots(runner_one, "OneZone")

print("\nGeminiTwoZoneNoiseModel ...")
fids_two = run_shots(runner_two, "TwoZone")


# ── Plot ──────────────────────────────────────────────────────────────────
labels = ["Ideal\n(noiseless)", "GeminiOne\nZoneNoise", "GeminiTwo\nZoneNoise"]
means  = [np.mean(fids_ideal), np.mean(fids_one), np.mean(fids_two)]
stds   = [np.std(fids_ideal),  np.std(fids_one),  np.std(fids_two)]
colors = ["C0", "C1", "C2"]

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(labels, means, yerr=stds, capsize=6,
              color=colors, alpha=0.85, edgecolor="k", linewidth=0.6)
for bar, m, s in zip(bars, means, stds):
    ax.text(
        bar.get_x() + bar.get_width() / 2, m + s + 0.003,
        f"{m:.4f}", ha="center", va="bottom", fontsize=9,
    )
ax.set_ylabel("Logical state fidelity  F(\u03c1, |+_L\u27e9)")
ax.set_title(
    f"Part 4 \u2014 Gemini One/Two Zone noise on Steane |+_L\u27e9 encoding\n"
    f"(shots = {GEMINI_SHOTS}, circuit: |0_L\u27e9 \u2192 H_L \u2192 |+_L\u27e9)"
)
ax.set_ylim(max(0.0, min(means) - max(stds) - 0.05), 1.02)
ax.axhline(1.0, color="gray", ls="--", lw=0.8, alpha=0.6)
ax.grid(True, axis="y", alpha=0.35)
fig.tight_layout()
plt.savefig(f"{RESULTS_DIR}/part4_gemini_noise.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"\nIdeal     : {np.mean(fids_ideal):.5f} \u00b1 {np.std(fids_ideal):.5f}")
print(f"OneZone   : {np.mean(fids_one):.5f} \u00b1 {np.std(fids_one):.5f}")
print(f"TwoZone   : {np.mean(fids_two):.5f} \u00b1 {np.std(fids_two):.5f}")
```

```python
# ============================================================
# Part 4 — Fault-tolerant |0_L⟩ preparation  (Goto 2016, Fig. 1c)
# ============================================================
#
# Goto 2016 protocol:
#   Non-FT:  encode |0_L⟩ with graph-state circuit (9 CZ = 9 CX gates).
#            Logical error rate  O(p_e)  per physical error.
#   FT:      Same non-FT encoding  +  1-ancilla Z-stabilizer verification:
#            Ancilla |0⟩, CNOT(data_i, anc) for i in {0,2,4,6},
#            then measure anc.  Accept iff anc = 0 (no X error detected).
#            Accepted shots: logical error rate  O(p_e^2).
#
# Z-error degeneracy:  Z_L|0_L⟩ = |0_L⟩  => phase errors are harmless.
# Only X errors matter.  Z_0 Z_2 Z_4 Z_6 catches odd-parity X errors on
# qubits {0,2,4,6} non-destructively.
#
# Noise: depolarize2 after every CX in encoding (9) + check (4, FT only).
# ============================================================

import importlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bloqade import squin
from bloqade.pyqrack import StackMemorySimulator

import gate_syntesis_helpers as gh
import gate_syntesis
importlib.reload(gh)
importlib.reload(gate_syntesis)


# -- Factory: noisy |0_L> encoding (depolarize2 after each CX in 9 CZ gates) -
def make_noisy_steane_zero(p2q: float):

    @squin.kernel
    def _noisy_cz(a, b):
        squin.h(b)
        squin.cx(a, b)
        squin.depolarize2(p=p2q, control=a, target=b)
        squin.h(b)

    @squin.kernel
    def _noisy_zero():
        q = squin.qalloc(7)
        for i in range(7):
            squin.h(q[i])
        _noisy_cz(q[0], q[6])
        _noisy_cz(q[1], q[3])
        _noisy_cz(q[5], q[3])
        _noisy_cz(q[0], q[4])
        _noisy_cz(q[5], q[6])
        _noisy_cz(q[1], q[2])
        _noisy_cz(q[0], q[2])
        _noisy_cz(q[5], q[4])
        _noisy_cz(q[1], q[4])
        squin.h(q[2]); squin.h(q[3]); squin.h(q[4]); squin.h(q[6])
        return q

    return _noisy_zero


# -- Non-FT kernel factory ---------------------------------------------------
def make_nFT_kernel(noisy_zero_fn):
    @squin.kernel
    def _nFT():
        return noisy_zero_fn()
    return _nFT


# -- FT kernel factory (encoding + Z0 Z2 Z4 Z6 stabilizer check) ------------
# Returns 8 qubits: data[0:7] + ancilla[7] (measured -> definite |0> or |1>).
def make_FT_kernel(noisy_zero_fn, p2q: float):

    @squin.kernel
    def _FT():
        q   = noisy_zero_fn()    # 7 data qubits with noise
        anc = squin.qalloc(1)    # ancilla |0>
        # Non-destructive Z0 Z2 Z4 Z6 stabilizer measurement:
        #   CNOT(data_i, anc) for i in {0,2,4,6}, then measure anc.
        #   anc=0 -> even-parity X errors on {0,2,4,6} (accept).
        squin.cx(q[0], anc[0]); squin.depolarize2(p=p2q, control=q[0], target=anc[0])
        squin.cx(q[2], anc[0]); squin.depolarize2(p=p2q, control=q[2], target=anc[0])
        squin.cx(q[4], anc[0]); squin.depolarize2(p=p2q, control=q[4], target=anc[0])
        squin.cx(q[6], anc[0]); squin.depolarize2(p=p2q, control=q[6], target=anc[0])
        squin.measure(anc[0])    # ancilla collapses to |0> or |1>
        return q + anc           # IList[Qubit, 8]: data[0:7] + ancilla[7]

    return _FT


# -- Logical zero state and fidelity -----------------------------------------
def _make_zero_L():
    psi = np.zeros(2**7, dtype=complex)
    for a in (0, 1):
        for b in (0, 1):
            for c in (0, 1):
                bits = [a, b, a^b, c, a^c, b^c, a^b^c]
                idx  = sum(bit << (6 - i) for i, bit in enumerate(bits))
                psi[idx] += 1.0 / np.sqrt(8)
    return psi

ZERO_L_VEC = _make_zero_L()

def _fid_zero(rho_128):
    return float(np.clip(np.real(np.vdot(ZERO_L_VEC, rho_128 @ ZERO_L_VEC)), 0.0, 1.0))

def _partial_trace_ancilla(rho_256):
    """Trace out ancilla (qubit 7 = LSB) from 256x256 DM.
    Qubit order: data=q[0:7] (MSBs), ancilla=q[7] (LSB).
    Returns (rho_data 128x128, rho_anc 2x2)."""
    r        = rho_256.reshape(128, 2, 128, 2)
    rho_data = np.trace(r, axis1=1, axis2=3)   # sum over anc  -> 128x128
    rho_anc  = np.trace(r, axis1=0, axis2=2)   # sum over data -> 2x2
    return rho_data, rho_anc


# -- Simulation --------------------------------------------------------------
SHOTS_FT = 20
P_VALUES  = [0.0, 0.002, 0.005, 0.01, 0.02, 0.05]

rows = []
print(f"{'p_noise':>8}  {'F_nFT':>12}  {'F_FT':>12}  {'accepted':>12}")
print("-" * 52)

for p in P_VALUES:
    noisy_zero = make_noisy_steane_zero(p)
    nFT_k      = make_nFT_kernel(noisy_zero)
    FT_k       = make_FT_kernel(noisy_zero, p)

    fids_nFT = []
    fids_FT  = []
    n_accept = 0

    for _ in range(SHOTS_FT):

        # Non-FT: 7 data qubits
        sim  = StackMemorySimulator(min_qubits=7)
        d    = sim.run(nFT_k)
        rho  = sim.reduced_density_matrix(d)
        fids_nFT.append(_fid_zero(rho))

        # FT: 8 qubits (data[0:7] + ancilla[7])
        sim2  = StackMemorySimulator(min_qubits=8)
        all_q = sim2.run(FT_k)
        rho_8 = sim2.reduced_density_matrix(all_q)     # 256x256

        rho_data, rho_anc = _partial_trace_ancilla(rho_8)
        anc_zero = rho_anc[0, 0].real > 0.5

        if anc_zero:
            n_accept += 1
            fids_FT.append(_fid_zero(rho_data))

    F_nFT = float(np.mean(fids_nFT))
    F_FT  = float(np.mean(fids_FT))  if fids_FT else float('nan')
    s_nFT = float(np.std(fids_nFT))
    s_FT  = float(np.std(fids_FT))   if fids_FT else float('nan')
    acc   = n_accept / SHOTS_FT

    rows.append(dict(p_noise=p, F_nFT=F_nFT, s_nFT=s_nFT,
                     F_FT=F_FT, s_FT=s_FT, accept_rate=acc, n_accepted=n_accept))
    print(f"{p:>8.3f}  {F_nFT:>7.4f}+/-{s_nFT:.4f}  "
          f"{F_FT:>7.4f}+/-{s_FT:.4f}  {n_accept:>3}/{SHOTS_FT} ({acc:.0%})")

df_ft = pd.DataFrame(rows)


# -- Plots -------------------------------------------------------------------
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(13, 4.5))

ax0.errorbar(df_ft['p_noise'], df_ft['F_nFT'], yerr=df_ft['s_nFT'],
             marker='o', capsize=4, color='C1', label='Non-FT encoding')
ax0.errorbar(df_ft['p_noise'], df_ft['F_FT'],  yerr=df_ft['s_FT'],
             marker='s', capsize=4, color='C0', linestyle='--',
             label='FT + Z0Z2Z4Z6 check (Goto Fig 1c, accepted shots)')
ax0.set_xlabel('Physical CX error rate  p2q  (per gate)')
ax0.set_ylabel('Logical fidelity  F(rho, |0_L>)')
ax0.set_title('Non-FT vs FT |0_L> preparation')
ax0.set_ylim(-0.05, 1.05)
ax0.legend(fontsize=9); ax0.grid(True)

ax1.bar(range(len(P_VALUES)), df_ft['accept_rate'],
        tick_label=[f'{p:.3f}' for p in P_VALUES],
        color='C2', alpha=0.8, edgecolor='k', linewidth=0.6)
ax1.axhline(1.0, color='gray', ls=':', lw=0.8, alpha=0.6)
ax1.set_xlabel('Physical CX error rate  p2q')
ax1.set_ylabel('FT acceptance rate')
ax1.set_title('FT post-selection rate\n(rejected = Z0Z2Z4Z6 syndrome detected)')
ax1.set_ylim(-0.05, 1.10)
ax1.grid(True, axis='y')

fig.suptitle(
    'Part 4 -- Fault-tolerant |0_L> preparation  (Goto 2016, Fig. 1c)\n'
    'Noise: depolarize2 on every CX.  Z-error degeneracy: Z_L|0_L> = |0_L> => only X errors harmful.\n'
    'FT checks Z0Z2Z4Z6 (4 extra noisy CX).  At low p: F_FT > F_nFT (O(p^2) vs O(p) logical errors).',
    fontsize=8,
)
fig.tight_layout()
plt.savefig(f"{RESULTS_DIR}/part4_ft_preparation.png", dpi=150, bbox_inches='tight')
plt.show()

display(df_ft[['p_noise', 'F_nFT', 's_nFT', 'F_FT', 's_FT', 'accept_rate']])
```
